## Data Exploration with Word2Vec

```python
from gensim.models.word2vec import Word2Vec
import pandas as pd
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
from viz_engine import viz_engine
from word_embedder import word_embedder
```

```python
file = pd.read_csv('assets/WikiLarge_Train.csv', encoding='utf-8')

SAMPLE_SIZE = 50000#len(file)

np.random.seed(911)
df = file.sample(SAMPLE_SIZE)
```

```python
word_embedder.word_embedder(df,'test_model')
```

```python
model = Word2Vec.load('test_model.bin')
```

```python
model.wv.most_similar(['man'],topn=20)
```

```python
model.wv.most_similar(positive=['king','man'], negative = ['queen'])
```
